```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets, svm
```

```python
# X, y = datasets.load_iris(return_X_y=True)
# X, y = datasets.load_breast_cancer(return_X_y=True)
df = pd.read_csv('diabetes.csv')
y = df['Outcome']
X = df.drop(columns='Outcome')
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, random_state=0)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn.model_selection import cross_val_score, GridSearchCV
```

```python
svc = svm.SVC(kernel='linear', C=100, random_state=42)
scores = cross_val_score(svc, X, y, cv=5)
scores.mean(), scores.std()
```

```python
scores
```

```python
param_grid = [
    {'C': [0.1, 1, 10, 100, 1000], 'kernel': ['linear']},
     {'C': [0.1, 1, 10, 100, 1000], 'gamma': [0.01, 0.001, 0.0001], 'kernel': ['poly', 'rbf']},
    ]
clf = GridSearchCV(svc, param_grid).fit(X, y)
clf.best_params_, clf.best_score_
```

```python
pd.DataFrame(clf.cv_results_)[['param_C', 'param_kernel', 'param_gamma', 'mean_test_score', 'std_test_score', 'rank_test_score']]
```
