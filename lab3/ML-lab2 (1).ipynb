```python
df = pd.read_csv("reglin_1.csv")
df
```

```python
df.info()
```

```python
plt.plot(df['x'], df['y'], '.');
plt.title('reglin_1');
```

```python
n = len(df)
Sx = df['x'].sum()
Sy = df['y'].sum()
Sxy = (df['x'] * df['y']).sum()
# Sxy = df['x'].dot(df['y'])
Sx2 = (df['x']**2).sum()
# Sx2 = np.square(df['x']).sum()
# Sx2 = df['x'].dot(df['x'])
```

```python
W = n * Sx2 - Sx**2
Wa = n * Sxy - Sx * Sy
Wb = Sx2 * Sy - Sx * Sxy
```

```python
a = Wa/W
b = Wb/W
a, b
```

```python
xl = np.linspace(-0.5, 10.5, 500)
yl = a*xl + b
```

```python
plt.plot(df['x'], df['y'], '.');
plt.plot(xl, yl, 'r-');
plt.title('reglin_1');
```

```python
rss = np.square(df['y'] - a*df['x'] - b).sum()
tss = np.square(df['y'] - df['y'].mean()).sum()
r2 = 1.0 - rss/tss
np.sqrt(rss/tss), r2
```

```python
xl = np.linspace(-0.5, 10.5, 500)
# xl[:, np.newaxis].shape
xl.reshape(-1, 1).shape
# fit() and predict() both require X as a matrix where each row is one sample, and columns are values
```
