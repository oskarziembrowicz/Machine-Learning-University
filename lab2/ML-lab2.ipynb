```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("reglin_1.csv")
df
```

```python
df.info()
```

```python
plt.plot(df['x'], df['y'], '.');
```

```python
n = len(df)
Sx = df['x'].sum()
Sy = df['y'].sum()
Sxy = (df['x'] * df['y']).sum()
# Sxy = df['x'].dot(df['y'])
Sx2 = (df['x']**2).sum()
# Sx2 = np.square(df['x']).sum()
# Sx2 = df['x'].dot(df['x'])
```

```python
W = n * Sx2 - Sx**2
Wa = n * Sxy - Sx * Sy
Wb = Sx2 * Sy - Sx * Sxy
```

```python
a = Wa/W
b = Wb/W
a, b
```

```python
xl = np.linspace(-0.5, 10.5, 500)
yl = a*xl + b
```

```python
plt.plot(df['x'], df['y'], '.');
plt.plot(xl, yl, 'r-');
```
